## Modelling

```python
train.dropna(subset=["Embarked"],inplace=True)
train.isnull().sum()
```

```python
x_train = train[["Sex","Age","Pclass","SibSp","Parch","Embarked","Fare"]]
x_train
```

```python
y_train = train[["Survived"]]
y_train
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()

x_train['Sex']=le.fit_transform(x_train['Sex'])
x_train['Embarked']=le.fit_transform(x_train['Embarked'])
x_train
```

```python
test.isna().sum()
```

```python
test['Title'] = test.Name.str.extract('([A-Za-z]+)\.')
test["Title"].replace(Mn,'Men',inplace=True)
test["Title"].replace(Wn,'Women',inplace=True)
test["Title"].replace(boy,'Boy',inplace=True)
test["Title"].replace(girl,'Girl',inplace=True)

medianmen= np.median(test[test["Title"]=='Men']['Age'].dropna())
test.loc[(test['Title']=='Men') & (test['Age'].isna()),'Age']  = medianmen

medianwomen= np.median(test[test["Title"]=='Women']['Age'].dropna())
test.loc[(test['Title']=='Women') & (test['Age'].isna()),'Age']  = medianwomen

medianboy= np.median(test[test["Title"]=='Boy']['Age'].dropna())
test.loc[(test['Title']=='Boy') & (test['Age'].isna()),'Age']  = medianboy

mediangirl= np.median(test[test["Title"]=='Girl']['Age'].dropna())
test.loc[(test['Title']=='Girl') & (test['Age'].isna()),'Age']  = mediangirl
test.isna().sum()
```

```python
#test.dropna(subset=["Fare"],inplace=True)
#test.isnull().sum()
```

```python
x_test=test[["Sex","Age","Pclass","SibSp","Parch","Embarked","Fare"]]
y_test=test[["Survived"]]
x_test['Sex']=le.fit_transform(x_test['Sex'])
x_test['Embarked']=le.fit_transform(x_test['Embarked'])
x_test.describe()
```

```python
x_test[x_test.isnull().any(axis=1)]
x_test["Fare"].fillna(7.895800,inplace=True)
x_test.isna().sum()
```

## DECISON TREE

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV
#classifier= DecisionTreeClassifier(criterion='entropy', random_state=0,max_depth=5)  

parameters = {'max_depth':range(3,30)}
classifier = GridSearchCV(DecisionTreeClassifier(), parameters, n_jobs=4)
classifier.fit(X=x_train, y=y_train)
tree_model = classifier.best_estimator_
print (classifier.best_score_, classifier.best_params_) 
```

```python
#classifier.fit(x_train, y_train)
y_pred= tree_model.predict(x_test)
```

```python
from sklearn import metrics
cm= metrics.confusion_matrix(y_test, y_pred)
cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix = cm, display_labels = [0, 1])

cm_display.plot()
plt.show()
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
pred = pd.DataFrame(y_pred)
pred.to_csv("DecisionTreeFare.csv")
```

## Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
#LR = LogisticRegression(solver='liblinear')
parameters = {"C":[0.001, 0.01, 0.1, 1, 10, 100], "penalty":["l1","l2"], "solver":['newton-cg', 'lbfgs', 'liblinear', 'sag']}
LR = GridSearchCV(LogisticRegression(), parameters, n_jobs=4)
LR.fit(X=x_train, y=y_train)

print("tuned hpyerparameters :(best parameters) ",LR.best_params_)
print("accuracy :",LR.best_score_)
```

```python
#LR = LogisticRegression(C=0.1,solver='liblinear').fit(x_train,y_train)
y_pred= LR.predict(x_test)
```

```python
from sklearn import metrics
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
from sklearn import metrics
lr= metrics.confusion_matrix(y_test, y_pred)
lr_display = metrics.ConfusionMatrixDisplay(confusion_matrix = lr, display_labels = [0, 1])

lr_display.plot()
plt.show()
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
pred = pd.DataFrame(y_pred)
pred.to_csv("NewtonFare.csv")
```

## RandomForest

```python
from sklearn.ensemble import RandomForestClassifier  
RF= RandomForestClassifier(criterion='gini', 
                             n_estimators=100,
                             min_samples_split=3,
                             min_samples_leaf=1,
                             max_features='auto',
                             oob_score=True,
                             random_state=1,
                             n_jobs=4)  
RF.fit(x_train, y_train)
y_pred= RF.predict(x_test)
```

```python
from sklearn import metrics
rf= metrics.confusion_matrix(y_test, y_pred)
rf_display = metrics.ConfusionMatrixDisplay(confusion_matrix = rf, display_labels = [0, 1])

rf_display.plot()
plt.show()
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
pred = pd.DataFrame(y_pred)
pred.to_csv("RandomForestFare.csv")
```
